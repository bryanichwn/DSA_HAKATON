## Importing the Libraries and Dataset

```python
#Import the libraries
import pandas as pd 
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Slicing the Dataset
df = df.iloc[:,:5]
df.head()
```

```python
#Descriptive Statistics
df.describe()
```

```python
#Time Series Plot
columns = ['Positif (Indonesia)', 'Dirawat (Indonesia)',
       'Sembuh (Indonesia)', 'Meninggal (Indonesia)']
for i in columns:
    plt.plot(df['Tanggal'],df[i])
    plt.xlabel('Tanggal')
    plt.ylabel(i+" --> (orang)")
    plt.xticks(rotation=45)
    plt.show()
```

## Merging The Datasets

```python
#Import Mobility Data 2020
mobil = pd.read_csv("Mobility_Indonesia.csv")
mobil.head()
```

```python
mobil.info()
```

```python
#Changing the Date Datatype
mobil["date"] = mobil["date"].astype("datetime64")
```

```python
mobil.info()
```

```python
#Import Mobility Data 2021
mobil2 = pd.read_csv("Mobility_Indonesia_2021.csv")
mobil2.head()
```

```python
#Changing the Date Datatype
mobil2["date"] = mobil2["date"].astype("datetime64")
```

```python
#Concating the Mobility Datasets
data_me = pd.concat([mobil,mobil2],axis=0)
data_me.head()
```

```python
#Merging the Datasets on Date
data_merge = pd.merge(df,data_me,how="left",left_on="Tanggal",right_on = "date")
```

```python
data_merge.head()
```

```python
#Identifying the Number of Null Values
data_merge["sub_region_1"].isnull().sum()
```

```python
#Selecting Indonesian Mobility Data Only
data_merge = data_merge[data_merge["sub_region_1"].isnull()]
data_merge.head()
```

```python
data_merge.info()
```

```python
#Drop Unneccesary Columns
data_merge = data_merge.drop(data_merge[data_merge.columns[5:13]],axis=1)
```

```python
data_merge.head()
```

```python
#Dropping Null Values
data_merge = data_merge.dropna(axis=0)
```

```python
data_merge.head()
```

```python
data_merge = data_merge.drop("date",axis=1)
```

```python
data_merge.head()
```

```python
#Changing the Index to Date
data_merge = data_merge.set_index("Tanggal")
```

```python
data_merge.head()
```

```python
#Importing the Positivity Rate Data
df2 = pd.read_excel("covid_indonesia.xlsx",sheet_name = "Hasil Lab",index_col='Tanggal')
pos_data = df2[['Total Spesimen di Test','Positif']]
pos_data['Positivity Rate'] = pos_data['Positif']/pos_data['Total Spesimen di Test']
```

```python
#Merging the Datasets
data_merge = pd.merge(data_merge,pos_data,left_on='Tanggal',right_on=pos_data.index)
```

```python
data_merge.drop(['Positif (Indonesia)', 'Dirawat (Indonesia)','Sembuh (Indonesia)', 'Meninggal (Indonesia)','Total Spesimen di Test','Positif'],axis=1,inplace=True)
```

```python
#Normal Positivity Rate
columns = ['retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline',
       'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline','workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']

fig, ((ax1,ax2),(ax3,ax4), (ax5,ax6)) = plt.subplots(3,2,figsize=(16,16))

for ax,col in zip(fig.axes, columns):
    axsec = ax.twinx()
    ax.plot(data_merge['Tanggal'],data_merge[col])
    axsec.plot(data_merge['Tanggal'],data_merge['Positivity Rate'],color='orange')
    ax.set_xlabel('Tanggal')
    ax.legend(['Mobility'],loc='upper right')
    axsec.legend(['Positive Rate'],loc=(0.785,0.86))
    ax.set_ylabel(col)
    for tick in ax.get_xticklabels():
        tick.set_rotation(45)
        
fig.tight_layout()
    
```

## Outlier Handling

```python
#Box Plot
sns.boxplot(df_shifted['Positivity Rate'])
plt.show()
```

```python
df_shifted.head()
```

```python
#Mean Imputation
IQR = df_shifted['Positivity Rate'].quantile(0.75) - df_shifted['Positivity Rate'].quantile(0.25)
rangevalue = 1.5*IQR
Minvalue = df_shifted['Positivity Rate'].quantile(0.25)-rangevalue
Maxvalue = df_shifted['Positivity Rate'].quantile(0.75)+rangevalue
df_shifted['Positivity Rate'][df_shifted['Positivity Rate'] < Minvalue] = np.mean(df_shifted['Positivity Rate'])
df_shifted['Positivity Rate'][df_shifted['Positivity Rate'] > Maxvalue] = np.mean(df_shifted['Positivity Rate'])
```

```python
columns = ['retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline',
       'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline','workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']

fig, ((ax1,ax2),(ax3,ax4), (ax5,ax6)) = plt.subplots(3,2,figsize=(16,16))

for ax,col in zip(fig.axes, columns):
    ax.scatter(df_shifted[col],df_shifted['Positivity Rate'])
    ax.set_xlabel(col)
    ax.set_ylabel('Positivity Rate')
```

## Preprocessing

```python
#Variable Statement
X = df_shifted.drop(['Positivity Rate'],axis = 1)
y = df_shifted[['Positivity Rate']]
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Standard Scaling
from sklearn.preprocessing import StandardScaler
xscaler = StandardScaler()
X_train = xscaler.fit_transform(X_train)
X_test = xscaler.transform(X_test)

yscaler = StandardScaler()
y_train = yscaler.fit_transform(y_train)
y_test = yscaler.transform(y_test)
```

### Linear Regression

```python
#Fitting the Model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
#R2 Score training
linreg.score(X_train,y_train)
```

```python
#RMSE testing
from sklearn.metrics import mean_squared_error
y_pred = linreg.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
#RMSE cross validation
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(LinearRegression(),X_train,y_train,cv=5))
```

```python
#Coefficients
coefs = linreg.coef_.reshape(-1)
features = X.columns
magnitude = pd.DataFrame({'Feature':features,'Coefficient':coefs})
magnitude
```

## Lasso Regression

```python
#Grid Search
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
params_grid = {'alpha':[0.0001,0.01,0.02, 0.024, 0.025, 0.026, 0.03]}
gcv = GridSearchCV(Lasso(),params_grid,cv=10,scoring='neg_mean_squared_error')
```

```python
#Fitting the Model
gcv.fit(X_train,y_train)
```

```python
#Best Hyperparameters
gcv.best_params_
```

```python
#Fitting the Model
lasso = Lasso(alpha=0.0001)
lasso.fit(X_train,y_train)
```

```python
#R2 Score training
lasso.score(X_train,y_train)
```

```python
#RMSE testing
y_pred = lasso.predict(X_test)
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
#RMSE cross validation
from sklearn.model_selection import cross_val_score
np.sqrt(abs(np.mean(cross_val_score(Lasso(alpha=0.0001),X_train,y_train,cv=10,scoring='neg_mean_squared_error'))))
```

### Support Vector Regression

```python
#Grid Search
from sklearn.svm import SVR
svr = SVR(kernel='rbf')
params_grid = {'C': [0.01,0.1,1, 10], 'gamma': [1,0.1,0.01,0.001]}
gcv = GridSearchCV(SVR(kernel='rbf'),params_grid,cv=10,scoring='neg_mean_squared_error')
gcv.fit(X_train,y_train)
```

```python
#Best Hyperparameters
gcv.best_params_
```

```python
#Fitting the Model & RMSE Testing
svr = SVR(kernel='rbf',C=1,gamma=1)
svr.fit(X_train,y_train)
from sklearn.metrics import mean_squared_error
y_pred = svr.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
#R2 Score Training
svr.score(X_train,y_train)
```

```python
#RMSE Cross Validation
from sklearn.model_selection import cross_val_score
np.sqrt(abs(np.mean(cross_val_score(SVR(kernel='rbf',C=1,gamma=1),X_train,y_train,cv=10,scoring='neg_mean_squared_error'))))
```
